### Descriptive Statistics Calculator

```python
arr = np.array([1, 2, 2, 3, 4, 4, 4, 5])

arr
```

```python
print(np.mean(arr))
print(np.median(arr))
print(np.argmax(np.bincount(arr))) # Mode 
print(np.var(arr))
print(np.std(arr)) # or np.sqrt(np.var(arr))
```

```python
print(np.quantile(arr, 0.25))
print(np.quantile(arr, 0.5))
print(np.quantile(arr, 0.75))

IQR = np.quantile(arr, 0.75) - np.quantile(arr, 0.25)

IQR 
```

```python
# Getting mode if there is -ve number 

arr = np.array([-5, 0, 5, 10, 15])

negatives = arr[arr < 0]

if negatives.size > 0:
    values, counts = np.unique(negatives, return_counts=True)
else:
    values, counts = np.unique(arr, return_counts=True)

mode = values[np.argmax(counts)]

mode
```

### Calculate Covariance Matrix

```python
list_ = [[1, 2, 3], [4, 5, 6]] 
print(list_)
```

```python
len(list_[0]), len(list_)
```

```python
centerd = [sum(i)/len(list_[0]) for i in list_]
centerd
```

```python
cov_mat =  [[0]*len(list_) for _ in range(len(list_))]
cov_mat
```

```python
d = len(list_)
n = len(list_[0])

for i in range(d):
    for j in range(d):
        s = 0.0
        for k in range(n):
            s += (list_[i][k] - centerd[i]) * (list_[j][k] - centerd[j])
        cov_mat[i][j] = s / (n - 1)

print(cov_mat)
```
