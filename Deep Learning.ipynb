### Sigmoid Activation Function Understanding

The sigmoid function (also called the logistic function) is a commonly used activation function in machine learning and neural networks.

any real number (−∞,+∞) Output will be a value between 0 and 1



```python
# e is the Euler’s number, 

import math 


def get_sigmoid(number):

    euler_number = math.e

    return round(1 / (1 + (euler_number ** - number)), 4)

    print(euler_number)

print(get_sigmoid(0))
print(get_sigmoid(1))
```
