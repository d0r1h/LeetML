### Sampling Distribution of the Mean

Demonstrates the sampling distribution of the mean. The function should take as input the number of samples, the sample size, and the distribution type ('uniform' or 'exponential'). It should return the mean of the sample means.

The sampling distribution of the mean is a probability distribution of sample means, not of individual data points.



```python
import numpy as np

def simulate_clt(num_samples: int, sample_size: int, distribution: str = 'uniform') -> float:

    sample_means = []
    for _ in range(num_samples):
        if distribution == 'uniform':
            
            sample = np.random.uniform(0,1,sample_size)
            sample_means.append(np.mean(sample))

        else:

            sample = np.random.exponential(1,sample_size)
            sample_means.append(np.mean(sample))

    sample_means = np.array(sample_means)
    return round(np.mean(sample_means),4)

simulate_clt(100, 20, 'uniform')
```
