### Grayscale Image Contrast Calculator

```python
import numpy as np 
```

A Grayscale Image Contrast Calculator can help you evaluate the contrast of a grayscale image using various methods such as 

1. calculating the difference between maximum and minimum pixel values or 
2. using specific formulas for contrast enhancement. 

*  __Range-based Contrast__

A straightforward method for calculating contrast in a grayscale image involves the difference between the maximum and minimum pixel values.

Contrast=Max(I)âˆ’Min(I) : where I represents the pixel values in the image.

In grayscale images, contrast specifically relates to the range between the darkest and lightest pixel values.

```python
img = np.array([[0, 50], [200, 255]])
img
```

```python
max_ = max([j for i in img for j in i])
min_ = min([j for i in img for j in i])

max_, min_
```

```python
max_ - min_
```
